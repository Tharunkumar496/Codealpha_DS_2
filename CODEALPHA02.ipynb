```python
import zipfile
import pandas as pd

# Unzip the file
with zipfile.ZipFile('/content/archive (1).zip', 'r') as zip_ref:
    zip_ref.extractall('/content/unemployment_data')

# Assuming the unzipped file is a CSV, let's find it and load it.
# Often, zip files contain a single CSV or a few. We'll look for a common CSV name.
# If this doesn't work, we might need to inspect the contents of the unzipped directory.
import os

unzipped_files = os.listdir('/content/unemployment_data')
csv_file = [f for f in unzipped_files if f.endswith('.csv')]

if csv_file:
    df = pd.read_csv(os.path.join('/content/unemployment_data', csv_file[0]))
    print(f"Successfully loaded {csv_file[0]} into a DataFrame.")
    print("Displaying the first 5 rows:")
    display(df.head())
else:
    print("No CSV file found in the unzipped directory. Please check the contents manually.")
```

```python
df.columns = ['Region', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'Area']

# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Display data types and check for missing values
print("\nDataFrame Info:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDescriptive Statistics:")
display(df.describe())
```

```python
df.dropna(inplace=True)

print("\nDataFrame Info after dropping missing values:")
df.info()

print("\nMissing Values after dropping NaN rows:")
print(df.isnull().sum())

print("\nDescriptive Statistics after dropping NaN rows:")
display(df.describe())
```

```python
import plotly.express as px

# Group by date and calculate the mean unemployment rate across all regions
unemployment_rate_over_time = df.groupby('Date')['Estimated Unemployment Rate'].mean().reset_index()

fig = px.line(
    unemployment_rate_over_time,
    x='Date',
    y='Estimated Unemployment Rate',
    title='Estimated Unemployment Rate Over Time (All India Average)',
    labels={
        'Date': 'Date',
        'Estimated Unemployment Rate': 'Average Estimated Unemployment Rate (%)'
    },
    template='plotly_white'
)

fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
# Visualize unemployment rate by region
fig = px.line(
    df,
    x='Date',
    y='Estimated Unemployment Rate',
    color='Region',
    title='Estimated Unemployment Rate by Region Over Time',
    labels={
        'Date': 'Date',
        'Estimated Unemployment Rate': 'Estimated Unemployment Rate (%)',
        'Region': 'Region'
    },
    template='plotly_white'
)

fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
# Group by Region, Area, and Date to see trends for rural and urban areas separately
unemployment_by_area_region = df.groupby(['Region', 'Area', 'Date'])['Estimated Unemployment Rate'].mean().reset_index()

fig = px.line(
    unemployment_by_area_region,
    x='Date',
    y='Estimated Unemployment Rate',
    color='Region',
    line_dash='Area', # Differentiate rural/urban with line style
    title='Estimated Unemployment Rate by Region and Area Over Time',
    labels={
        'Date': 'Date',
        'Estimated Unemployment Rate': 'Estimated Unemployment Rate (%)',
        'Region': 'Region',
        'Area': 'Area'
    },
    template='plotly_white'
)

fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
# Define pre and post COVID-19 periods (assuming March 2020 as the starting point for COVID impact)
pre_covid_df = df[df['Date'] < '2020-03-01']
post_covid_df = df[df['Date'] >= '2020-03-01']

# Calculate average unemployment rate for each period
avg_unemployment_pre_covid = pre_covid_df['Estimated Unemployment Rate'].mean()
avg_unemployment_post_covid = post_covid_df['Estimated Unemployment Rate'].mean()

print(f"Average Estimated Unemployment Rate (Pre-COVID): {avg_unemployment_pre_covid:.2f}%")
print(f"Average Estimated Unemployment Rate (Post-COVID): {avg_unemployment_post_covid:.2f}%")

# Visualize the distribution of unemployment rates before and after COVID-19
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2, subplot_titles=('Pre-COVID Unemployment Rate Distribution', 'Post-COVID Unemployment Rate Distribution'))

fig.add_trace(go.Histogram(x=pre_covid_df['Estimated Unemployment Rate'], name='Pre-COVID', marker_color='#636EFA'), row=1, col=1)
fig.add_trace(go.Histogram(x=post_covid_df['Estimated Unemployment Rate'], name='Post-COVID', marker_color='#EF553B'), row=1, col=2)

fig.update_layout(title_text='Distribution of Unemployment Rates Before and After March 2020', showlegend=False)
fig.show()
```

```python
# Calculate regional average unemployment rates for pre and post-COVID periods
regional_unemployment_pre_covid = pre_covid_df.groupby('Region')['Estimated Unemployment Rate'].mean().reset_index()
regional_unemployment_post_covid = post_covid_df.groupby('Region')['Estimated Unemployment Rate'].mean().reset_index()

# Merge the two dataframes
regional_unemployment_comparison = pd.merge(regional_unemployment_pre_covid,
                                              regional_unemployment_post_covid,
                                              on='Region',
                                              suffixes=('_pre_covid', '_post_covid'))

# Calculate the percentage change
regional_unemployment_comparison['Percentage_Change'] = ((regional_unemployment_comparison['Estimated Unemployment Rate_post_covid'] -
                                                         regional_unemployment_comparison['Estimated Unemployment Rate_pre_covid']) /
                                                        regional_unemployment_comparison['Estimated Unemployment Rate_pre_covid']) * 100

# Sort by percentage change to see which regions were most affected
regional_unemployment_comparison_sorted = regional_unemployment_comparison.sort_values(by='Percentage_Change', ascending=False)

print("\nRegional Unemployment Rate Changes (Pre vs. Post-COVID):")
display(regional_unemployment_comparison_sorted.round(2))

# Visualize the percentage change in unemployment rate by region
fig = px.bar(
    regional_unemployment_comparison_sorted,
    x='Region',
    y='Percentage_Change',
    title='Percentage Change in Estimated Unemployment Rate by Region (Pre vs. Post-COVID)',
    labels={
        'Region': 'Region',
        'Percentage_Change': 'Percentage Change (%)'
    },
    template='plotly_white'
)
fig.show()
```
